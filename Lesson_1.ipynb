__1.__ Найдите скалярное произведение векторов $x, y \in \mathbb{R}$:<br>
а) $x=(0,-3, 6),~y=(-4, 7, 9);$<br>
б) $x=(7, -4, 0, 1),~y=(-3, 1, 11, 2).$



```python
import numpy as np
xa = np.array([0,-3,6])
ya = np.array([-4,7,9])
xb = np.array([7,-4,0,1])
yb = np.array([-3,1,11,2])

print(f"Ответ:\n a) {xa @ ya}\n б) {xb @ yb}")
# скалярное произведение это когда берётся 1 координата из х и y, перемножаются и складываются с остальными координатами
# для a) это будет 0*-4 + -3*7 + 6*9 = 54-21 = 33
```

__2.__ Найдите нормы векторов $(4, 2, 4)$ и $(12, 3, 4)$ и угол между ними.




```python
#манхэтеннская считается как сумма модулей координат вектора
# а Евклидова считается как корень из квадратов координат вектора
from numpy.linalg import norm
a = np.array([4,2,4])
b = np.array([12,3,4])
print(f"Манхэттенская норма вектора a = {norm(a, ord=1)}\nЕвклидова норма вектора a = {norm(a, ord=2)}")
print()
print(f"Манхэттенская норма вектора b = {norm(b, ord=1)}\nЕвклидова норма вектора b = {norm(b, ord=2)}")
cos_phi = np.dot(a,b) / norm(a) / norm(b)
print()
print(f"Угол между векторами {cos_phi}")
```
